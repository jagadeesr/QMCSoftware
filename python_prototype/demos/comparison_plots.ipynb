```python
from pandas import read_csv
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline
```

```python
distrib_names = ['IIDStdUniform','IIDStdGaussian','Lattice','Sobol']
```

# Absolute Tolerance Plots

```python
df_abstols = read_csv('../outputs/comparison_tests/abs_tol.csv')
df_abstols.loc[:10]
```

```python
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
for distrib_name in distrib_names:
    ax[0].plot(df_abstols['abs_tol'],df_abstols[distrib_name+'_solution'], label=distrib_name)
    ax[1].loglog(df_abstols['abs_tol'],df_abstols[distrib_name+'_n'], label=distrib_name)
ax[0].legend(loc='upper left')
ax[0].set_xlabel('Absolute Tolerance')
ax[0].set_ylabel('Runtime')
ax[1].set_xlabel('Absolute Tolerance')
ax[1].set_ylabel('Total Samples')
fig.suptitle('Comparing Absolute Tolerances')
plt.savefig('../outputs/comparison_tests/abstols_plot.png',dpi=200)
```

# Dimension Plots

```python
df_dimensions = read_csv('../outputs/comparison_tests/dimension.csv')
df_dimensions.loc[:10]
```

```python
fig,ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))
for distrib_name in distrib_names:
    ax[0].plot(df_dimensions['dimension'],df_dimensions[distrib_name+'_solution'], label=distrib_name)
    ax[1].loglog(df_dimensions['dimension'],df_dimensions[distrib_name+'_n'], label=distrib_name)
ax[0].legend(loc='upper right')
ax[0].set_xlabel('Dimensions')
ax[0].set_ylabel('Runtime')
ax[1].set_xlabel('Dimensions')
ax[1].set_ylabel('Total Samples')
fig.suptitle('Comparing Dimensions')
plt.savefig('../outputs/comparison_tests/dimension_plot.png',dpi=200)
```
